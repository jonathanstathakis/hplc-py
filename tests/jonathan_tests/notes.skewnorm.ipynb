# Standard Normal Distribution

The normal distribution is as follows:

$$
\phi(x) = \frac{1}{\sqrt{2 \pi}}e^{-\frac{x^2}{2}}
$$

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

x = np.linspace(-3,3,100)
y = stats.norm(0,1).pdf(x)

# print(x, y)
plt.plot(x,y)
plt.suptitle('Standard Normal Distribution')
plt.show()
```

Where the 'standard' refers to the special case of the distribution where $\mu=0$ and $\sigma=1$.

This distribution is described by the Probability Density Function (PDF), also known as the density:
$$
\phi(x)=\frac{
  e^{
    -x^2 / 2
  }
  }
  {
    \sqrt{
      2 \pi
      }
  }
$$

where a PDF describes the likelihood of sample space equaling the variable input value at that point. The output of the PDF is the probability per unit length. Note that while the absolute likelihood of a value to take on the value in the sample space is zero BECAUASE there is an infinite set of possible values, the PDF describes the probability of the variable to fall within a range of values.

The probability is expressed as the integral of the PDF over the specified range.

PDFs are nonnegative everywhere and area under the curve is equal to 1.

# CDF

The CDF of a PDF is the integral of that function over the range $- \infty \rightarrow x$

The cumulative distribution function describes the probability of a random value to have a value less than or equal to x. The CDF of the Standard Normal Distribution (SND) is often encountered as the z-table.

The CDF of the SND is:

$$
F(x;\mu,\sigma)=\frac{
  1
}
{
  \sigma \sqrt{
    2 \pi
  }
}
\int ^ x _{- \infty} \exp \left (
  \frac{
    (t - \mu ^ 2)
  }
  {
    2 \sigma ^2
  }
\right )
  dt
$$

or expressed as:

$$
\Phi (x) = \int ^x _{- \infty} \phi (t) dt = \frac{1}{2} \left [  1 + \text{erf} (\frac{x}{\sqrt 2}) \right ]
$$

# PDF of the Skew Normal Distribution

Given the above definition of the CDF of the SND, if we are to include a skew parameter $\alpha$, then the PDF of the standard skew-normal distribution (SSND) is:$$\Phi (x) = 2 \phi (x) \Phi ( \alpha x)$$

where $\alpha$ is known as the 'shape' parameter. For $\alpha=0$ the distribution reverts back to the SND. Increasing $alpha$ increases the skewness. As $\alpha$ approaches $\infty$ the density converges to the half-normal density function. The polarity of $\alpha$ determines which side of the vertical axis the density appears.
